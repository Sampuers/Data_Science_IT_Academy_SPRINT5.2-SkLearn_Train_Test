```python
import pandas as pd


df = pd.read_csv('housing data.csv',header=None)
df
```

```python
df.columns =['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head(20)
```

Selecciono la variable dependiente y la independiente, queremos predecir el valor medio de las viviendas en consecuencia del resto de características. 

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('MEDV', axis=1) # Serían las características
y = df['MEDV'] # Variable dependiente valor medio de las viviendas.
```

Dividimos la base de datos en modo train-test, con las variables x, y asignadas anteriormente. 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### - Exercici 2
##### Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).



Estandarizo los datos de las características numéricas para centrarlas en cero y escalarlas para que tengan una desviación estándar de uno.

```python
# Lo ejecuto para x_train

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Standarized = pd.DataFrame(scaler.fit_transform(x_train),columns=data_train_x.columns)
```

```python
Standarized
```

```python
# Lo ejecuto para x_test

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Standarized_test = pd.DataFrame(scaler.fit_transform(x_test),columns=data_test_x.columns)
```

```python
Standarized_test
```

```python
# Creación de características polinómicas para Standarized
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, interaction_only=True)

poly_train = pd.DataFrame(poly.fit_transform(Standarized),columns=poly.get_feature_names_out(Standarized.columns))
poly_train.head()
```

```python
# Creación de características polinómicas para Standarized_test
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, interaction_only=True)

poly_test = pd.DataFrame(poly.fit_transform(Standarized_test),columns=poly.get_feature_names_out(Standarized_test.columns))
poly_test.head()
```

No puedo hacer dummies porque existían 2 variables categóricas Chas y Ras y ya se han creado como numéricas.
